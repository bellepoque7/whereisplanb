```python
import matplotlib.pyplot as plt
def graph(x= [1,2,3,4], y = [2,4,6,8]):
    plt.figure(figsize=(20,10))
    plt.plot(x,y)
    plt.ylim([0,max(y)])
    plt.xlim([min(x),max(x)])
    plt.xlabel('date')
    plt.ylabel('price/100g')
    plt.legend()
    plt.xticks(rotation=90)
    
    return plt.show
graph(my_df[my_df['상태(상품, 중품)'] == '중품']['날짜'],my_df[my_df['상태(상품, 중품)'] == '중품']['조회일자 가격']/20)
```
